```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_parquet('../data/reference/ibge_PR.parquet')
print(f"Dados carregados: {len(df)} municípios")
df.head()
```

```python
print(f"Registros: {len(df):,}")
print(f"Colunas: {df.shape[1]}")
print(f"Colunas: {df.columns.tolist()}")
```

```python
print(f"Colunas: {df.columns.tolist()}")
print(f"Tipos: {df.dtypes}")
df.describe(include='all')
```

```python
print(f"Valores nulos: {df.isnull().sum()}")
print(f"Duplicatas: {df.duplicated().sum()}")
```

```python
# Selecionar colunas úteis
df_clean = df[['codigo_ibge', 'nome', 'latitude', 'longitude']].copy()
df_clean.columns = ['codigo_municipio', 'nome_municipio', 'latitude', 'longitude']
df_clean['codigo_municipio'] = df_clean['codigo_municipio'].astype(str).str[:6]

```

```python
print(f"Total de municípios: {len(df_clean)}")
print(f"Municípios com coordenadas: {df_clean['latitude'].notna().sum()}")
print(df_clean.head())
```

```python
# Top municípios por nome
print(f"\n   • Primeiros 10 municípios (ordem alfabética):")
print(df_clean.sort_values('nome_municipio')[['codigo_municipio', 'nome_municipio']].head(10).to_string(index=False))

# Estatísticas de coordenadas
print(f"\n   • Coordenadas:")
print(f"      - Latitude: min {df_clean['latitude'].min():.4f}, max {df_clean['latitude'].max():.4f}")
print(f"      - Longitude: min {df_clean['longitude'].min():.4f}, max {df_clean['longitude'].max():.4f}")
```

```python
resumo = {
    'Total de municípios': len(df_clean),
    'Municípios com coordenadas': df_clean['latitude'].notna().sum(),
    'Colunas': df_clean.shape[1],
}

for chave, valor in resumo.items():
    print(f"   • {chave}: {valor}")
```

```python
# Salvar
df_clean.to_parquet('../data/processed/ibge_clean_PR.parquet')
```
